```python
import numpy as np

# Create working copy
df_clean = df.copy()

# Replace impossible values with NaN
df_clean.loc[df_clean["Age"] == 9, "Age"] = np.nan
df_clean.loc[df_clean["Training_Hours_Per_Week"] == 172, "Training_Hours_Per_Week"] = np.nan
df_clean.loc[df_clean["VO2max"] == -8.4, "VO2max"] = np.nan
df_clean.loc[df_clean["Body_Fat_Pct"] == 99.2, "Body_Fat_Pct"] = np.nan
```

```python
df_clean.isna().sum()
```

```python
#Hidden Duplication detection
duplicates = df_clean[df_clean.duplicated(
    subset=df_clean.columns.difference(["Athlete_ID"]),
    keep=False
)]

duplicates.sort_values("Athlete_Name")
```

```python
# Remove duplicates while keeping first occurrence
df_clean = df_clean.drop_duplicates(
    subset=df_clean.columns.difference(["Athlete_ID"]),
    keep="first"
)

df_clean.shape
```

```python
df.duplicated()
```

```python
#Missing medal handling
df_clean["Medal"] = df_clean["Medal"].fillna("None")
df_clean["Medal"].value_counts()
```

```python
# Check remaining missing values
df_clean.isna().sum()
```

```python
#Sport-Specific Median Imputation for Missing Performance Variables
# Age by sport
df_clean["Age"] = df_clean["Age"].fillna(
    df_clean.groupby("Sport")["Age"].transform("median")
)

# Training Hours by sport
df_clean["Training_Hours_Per_Week"] = df_clean["Training_Hours_Per_Week"].fillna(
    df_clean.groupby("Sport")["Training_Hours_Per_Week"].transform("median")
)

# Altitude by sport
df_clean["Altitude_Training_m"] = df_clean["Altitude_Training_m"].fillna(
    df_clean.groupby("Sport")["Altitude_Training_m"].transform("median")
)

# VO2max by sport
df_clean["VO2max"] = df_clean["VO2max"].fillna(
    df_clean.groupby("Sport")["VO2max"].transform("median")
)
```

```python
#Gender-and-Sport Specific Imputation for Body Fat Percentage
df_clean["Body_Fat_Pct"] = df_clean["Body_Fat_Pct"].fillna(
    df_clean.groupby(["Gender","Sport"])["Body_Fat_Pct"].transform("median")
)
```

```python
df_clean.isna().sum()
```

```python
#Handling MNAR Variables with Missingness Indicators
# Missingness flags
df_clean["Reaction_Time_Missing"] = df_clean["Reaction_Time_ms"].isna().astype(int)
df_clean["WorldCupPoints_Missing"] = df_clean["World_Cup_Points_Preseason"].isna().astype(int)

# Then sport-based median imputation for modeling usability
df_clean["Reaction_Time_ms"] = df_clean["Reaction_Time_ms"].fillna(
    df_clean.groupby("Sport")["Reaction_Time_ms"].transform("median")
)

df_clean["World_Cup_Points_Preseason"] = df_clean["World_Cup_Points_Preseason"].fillna(
    df_clean.groupby("Sport")["World_Cup_Points_Preseason"].transform("median")
)
```

```python
df_clean.isna().sum()
```

```python
# Overall fallback medians
df_clean["Reaction_Time_ms"] = df_clean["Reaction_Time_ms"].fillna(
    df_clean["Reaction_Time_ms"].median()
)

df_clean["World_Cup_Points_Preseason"] = df_clean["World_Cup_Points_Preseason"].fillna(
    df_clean["World_Cup_Points_Preseason"].median()
)
```

```python
df_clean.isna().sum()
```

```python
#Feature Engineering
# 1. Binary target for Task 3
df_clean["Medal_Won"] = (df_clean["Medal"] != "None").astype(int)

# 2. Experience + Training composite
df_clean["Experience_Index"] = (
    df_clean["Previous_Olympics"] * df_clean["Training_Hours_Per_Week"]
)

# 3. Physiological efficiency score
df_clean["VO2_BodyFat_Ratio"] = (
    df_clean["VO2max"] / df_clean["Body_Fat_Pct"]
)

# 4. Country support score
df_clean["GDP_Tradition_Score"] = (
    df_clean["Country_GDP_per_capita"] *
    df_clean["Winter_Sport_Tradition_Index"]
)
```
